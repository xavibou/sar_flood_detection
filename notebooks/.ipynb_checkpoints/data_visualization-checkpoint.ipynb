## Segmenting water with a threshold

```python
thr = 0.015
idx = 11
sar, dem, mask = test_dataloader.dataset[idx]

sar_seg = np.zeros_like(sar)
sar_seg[sar < thr] = 1

plot_example(sar, sar_seg, mask, sar_contrast1=8, sar_contrast2=14, dem_contrast=2, all_grey=True)
```

## Filter by connected components

```python
segmap = filter_by_region(sar_seg[0,:,:], thr=20)
#plt.imshow(segmap, cmap="Greys_r")

plot_example(sar, segmap[None, ...], mask, sar_contrast1=10, sar_contrast2=14, dem_contrast=2, all_grey=True)
```
